```python
import requests
import pandas as pd
import numpy as np
import folium
from folium.plugins import MiniMap
from urllib.parse import quote
import bs4
import selenium
from selenium.webdriver.common.by import By
from selenium import webdriver
import time
from tqdm import tqdm
```

```python
url = 'https://map.kakao.com/'
driver = webdriver.Chrome("C:/Users/Ji Hoon/Desktop/chromedriver")
driver.get(url)
```

```python
df = pd.read_csv("C:/Users/Ji Hoon/Desktop/서울시 모범업소.csv")
```

```python
#리뷰 크롤러
#왠진 모르겠는데 첫번째에서 한번, 23번에서 한번 막히는데 다시 진행해주면 크롤은 됌..
review_list = []
for i in tqdm(range(len(df["업소명"]))):
    a = driver.find_element(By.XPATH, '//*[@id="search.keyword.query"]')
    a.send_keys(df["업소명"][i])
    b = driver.find_element(By.XPATH,'//*[@id="search.keyword.submit"]')
    b.click()
    time.sleep(0.6)
    try:
        d = driver.find_element(By.XPATH,'//*[@id="info.search.place.list"]/li[1]/div[4]/a')
        d.click()
    except:
        g = driver.find_element(By.XPATH, '//*[@id="search.tab1"]/a')
        g.click()
        review_list.append([[],[],[]])
   
#리뷰창이 떴다
    time.sleep(1)
    try:
        driver.switch_to.window(driver.window_handles[1])
        small_review_list=[]
        for i in range(1,4):
            try:
                f= driver.find_element(By.XPATH,'//*[@id="mArticle"]/div[6]/div[2]/ul/li[{}]/div[3]/p/span'.format(i))
                small_review_list.append(f.text)
            except:
                small_review_list.append("")
        review_list.append(small_review_list)
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        g = driver.find_element(By.XPATH, '//*[@id="search.tab1"]/a')
        g.click()
    except:
        k = driver.find_element(By.XPATH,'//*[@id="search.tab1"]/a')
        k.click()
```
