```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.scatter(df['친절'],df['별점X총리뷰횟수'])
```

```python
df[df['친절']>800].index
# 이상치 인덱스 확인
```

```python
df_out=df.drop(189)
# 이상치 제거
```

```python
# 한글 문제
# matplotlit의 기본 폰트에서 한글 지원되지 않기 때문에
# matplotlib의 폰트 변경 필요
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':  # 맥OS 
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # 윈도우
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...  sorry~~~')
```

```python
title_font = {
    'fontsize':12,
    'fontweight':'bold'
}
ax=sns.lmplot(x="친절",y="별점X총리뷰횟수",data=df_out)
# 선형회귀선을 그으면 95% 신뢰구간이 자동으로 표현됨
plt.title('친절도와 (별점X총리뷰횟수)의 상관관계',fontdict=title_font,pad=10)
plt.xlabel('친절도')
plt.ylabel('별점X총리뷰횟수')
plt.scatter(df_out['친절'],df_out['별점X총리뷰횟수'],color='orange')
```

```python
title_font = {
    'fontsize':12,
    'fontweight':'bold'
}
ax=sns.lmplot(x="청결",y="별점X총리뷰횟수",data=df_out)
# 선형회귀선을 그으면 95% 신뢰구간이 자동으로 표현됨
plt.title('쳥결도와 (별점X총리뷰횟수)의 상관관계',fontdict=title_font,pad=10)
plt.xlabel('청결도')
plt.ylabel('별점X총리뷰횟수')
plt.scatter(df_out['청결'],df_out['별점X총리뷰횟수'])
```

```python
title_font = {
    'fontsize':12,
    'fontweight':'bold'
}
ax=sns.lmplot(x="신선한재료",y="별점X총리뷰횟수",data=df_out)
# 선형회귀선을 그으면 95% 신뢰구간이 자동으로 표현됨
plt.title('재료신선도와 (별점X총리뷰횟수)의 상관관계',fontdict=title_font,pad=10)
plt.xlabel('재료신선도')
plt.ylabel('별점X총리뷰횟수')
plt.scatter(df_out['신선한재료'],df_out['별점X총리뷰횟수'],color='green')
```
